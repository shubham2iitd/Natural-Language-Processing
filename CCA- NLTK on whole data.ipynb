```python
concord1_agg_token1 = []
for i in range(len(df_concord_1["column1"])):
    a = nltk.wordpunct_tokenize(df_concord_1["column1"][i])
    concord1_agg_token1 = concord1_agg_token1 + a

concord2_agg_token2 = []
for i in range(len(df_concord_2["column2"])):
    a = nltk.wordpunct_tokenize(df_concord_2["column2"][i])
    concord2_agg_token2 = concord2_agg_token2 + a

concord1_agg_text1 = nltk.Text(concord1_agg_token1)
concord2_agg_text2 = nltk.Text(concord2_agg_token2)

# concord_fdist1 is a lsit of "words" in "contexts of a concerned word" arranged by sequence of their frequency 

concord_fdist1 = nltk.FreqDist(concord1_agg_text1)
concord_fdist2 = nltk.FreqDist(concord2_agg_text2)
```

```python
concord_fdist1_lower = concord_fdist1.keys()
concord_fdist1_1 = concord_fdist1.keys()
concord_fdist1_lower = [x.lower() for x in concord_fdist1_lower]

concord_p = concord_fdist1_1
for i in range(len(concord_p)):
    concord_p[i] = concord_p[i].replace(concord_p[i],str(i)+"_"+"|"+concord_p[i])


concord_fdist2_lower = concord_fdist2.keys()
concord_fdist2_2 = concord_fdist2.keys()
concord_fdist2_lower = [x.lower() for x in concord_fdist2_lower]

concord_q = concord_fdist2_2
for i in range(len(concord_q)):
    concord_q[i] = concord_q[i].replace(concord_q[i],str(i)+"_"+"|"+concord_q[i])


# concord_diff is a list containg the uncommon words from the two sets, each having "words" that are in the context of a certain word.
    
concord_diff = set(concord_fdist2_lower).symmetric_difference(set(concord_fdist1_lower))
concord_diff = list(concord_diff)

concord_diff1 = []

concord_diff
```
